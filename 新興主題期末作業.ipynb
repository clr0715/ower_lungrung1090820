```python
# Solution 1:
import pandas as pd
import numpy as np
census = pd.read_csv('nationalcensus.csv')
# census = pd.read_csv('http://203.68.184.214/~most15/Outline/Summer2020/data/nationalcensus.csv')
census.head()
```

```python
census.dtypes
```

```python
# Solution 2:
# 第 0 列是文字描述應該去刪除!
census.drop(index=[0], inplace=True)

# 'population' 欄位是 object 型態，應該轉為 'int' 型態
census.population = census.population.astype(int)
print('總人口數是 {:,d} 人'.format(census.population.sum()))
```

```python
# Solution 3:
census.pivot_table(index='age', columns=['sex','marital_status'], values='population', aggfunc='sum')
```

```python
# Solution 4:
# 選定臺北市大安區
city1 = census[census.site_id=='臺北市大安區']
city1Result = city1.pivot_table(index='age', columns='marital_status', values='population', aggfunc='sum')

city1Result['人口數'] = city1Result.sum(axis=1)
city1Result['未婚率'] = city1Result['未婚'] / city1Result['人口數']           # 未婚率 = 未婚 / 人口數
city1Result['結婚率'] = city1Result['有偶_不同性別'] / city1Result['人口數']  # 結婚率 = 有偶_不同性別 / 人口數
city1Result['離婚率'] = city1Result['離婚_不同性別'] / city1Result['人口數']  # 離婚率 = 離婚_不同性別 / 人口數
city1Result['喪偶率'] = city1Result['喪偶_不同性別'] / city1Result['有偶_不同性別']

print('臺北市大安區')
city1Result
```

```python
# Solution 4-1:
# 選定屏東縣潮州鎮
city1 = census[census.site_id=='屏東縣潮州鎮']
city1Result = city1.pivot_table(index='age', columns='marital_status', values='population', aggfunc='sum')

city1Result['人口數'] = city1Result.sum(axis=1)
city1Result['未婚率'] = city1Result['未婚'] / city1Result['人口數']           # 未婚率 = 未婚 / 人口數
city1Result['結婚率'] = city1Result['有偶_不同性別'] / city1Result['人口數']  # 結婚率 = 有偶_不同性別 / 人口數
city1Result['離婚率'] = city1Result['離婚_不同性別'] / city1Result['人口數']  # 離婚率 = 離婚_不同性別 / 人口數
city1Result['喪偶率'] = city1Result['喪偶_不同性別'] / city1Result['有偶_不同性別']

print('屏東縣潮州鎮')
city1Result
```
